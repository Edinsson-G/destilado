# Visualización de resulatdos

```python
import torch
from datos import graficar
import copy
```

carga de datos

```python
ruta="resultados/results_long_train_ipc_50/"
etiquetas=torch.load(ruta+"etiquetas").detach().to("cpu")
imgs=torch.load(ruta+"imgs",map_location=torch.device("cpu"))
perdida=torch.load(ruta+"perdida")
#cantidad de outer_loop por cada iteración
ol_x_it=20
#cantidad total de iteraciones hechas
total_iteraciones=53
accuracy_entrenamiento=torch.empty((total_iteraciones,ol_x_it))#iteraciones x outer_loop
accuracy_testeo=copy.deepcopy(accuracy_entrenamiento)
for iteracion in range(total_iteraciones):
    for outer_loop in range(ol_x_it):
        [accuracy_entrenamiento[iteracion,outer_loop],accuracy_testeo[iteracion,outer_loop]]=torch.load(ruta+"metrics_outer_loop_"+str(outer_loop)+"iteracion_"+str(iteracion))
```

Graficar los datos de entrenamiento

```python
graficar(accuracy_entrenamiento,torch.tensor(range(total_iteraciones)),filas=7,columnas=8,figsize=(40,25))
```

Graficar métricas de testeo

```python
graficar(accuracy_testeo,torch.tensor(range(total_iteraciones)),filas=7,columnas=8,figsize=(40,25))
```
